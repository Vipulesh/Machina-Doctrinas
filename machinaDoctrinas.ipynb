```python
med5=Xtrain.groupby('Country ID')['Political Stability'].median()
```

# model

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
from xgboost import XGBRegressor
xgr=XGBRegressor(
    eta=0.1,
    max_leaves=400,
    max_depth=5,
    random_state=42,
    silent=True
)
xgr.fit(Xtrain, Ytrain)
```

```python
ypred=xgr.predict(Xtrain)
print(mse(Ytrain,ypred))
xgr.fit(Xtest,Ytest)
ypred1=xgr.predict(Xtest)
print(mse(Ytest,ypred1))
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(xgr,Xtest,Ytest,cv=10,scoring='neg_mean_squared_error' ))
print(cross_val_score(xgr,Xtrain,Ytrain,cv=10,scoring='neg_mean_squared_error' ))
```
